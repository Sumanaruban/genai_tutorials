```python
from langchain.chains import LLMMathChain
from langchain_openai import OpenAI

import os  # used to access filepaths
```

```python
api_key = os.environ['openai_api_key']
# api_key = 'Set your openai_api_key'

llm = OpenAI(temperature=0, api_key=api_key) # This is improtant. The temperature must be 0. No Imagination.
llm_math = LLMMathChain.from_llm(llm, verbose=True)
```

```python
llm_math.invoke("If I have 480 apples, how can I divide it to 4 people equally?")
```

```python
llm_math.invoke("What is 234 raised to the .738 power?")
```

## Example 2 - LLMCheckerChain

```python
from langchain.chains import LLMCheckerChain
from langchain_openai import OpenAI

llm = OpenAI(temperature=0.7)

text = "What type of mammal lays the biggest eggs?"

checker_chain = LLMCheckerChain.from_llm(llm, verbose=True)

checker_chain.run(text)
```

```python
from langchain.chains import LLMSummarizationCheckerChain
from langchain_openai import OpenAI

llm = OpenAI(temperature=0.7)

text = "What type of mammal lays the biggest eggs?"

checker_chain = LLMSummarizationCheckerChain.from_llm(llm, verbose=True, max_checks=2)

checker_chain.run(text)
```
