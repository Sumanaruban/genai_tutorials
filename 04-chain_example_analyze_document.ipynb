# Analyze a single long document

AnalyzeDocumentChain chain
1. takes a single document
2. splits it up
3. and runs it through a CombineDocumentsChain.

```python
from langchain.chains import AnalyzeDocumentChain
from langchain_openai import ChatOpenAI
from langchain.chains.question_answering import load_qa_chain
import os

# Settings the warnings to be ignored 
import warnings 
warnings.filterwarnings('ignore')

from IPython.display import Markdown
```

```python
# Read document
def read_document():
    with open("bbc.txt") as f:
        bbc_news = f.read()
    return bbc_news

bbc_news = read_document()
```

```python
# initialize LLM Model and create question answering document chain
api_key = os.environ['openai_api_key']
# api_key = 'Set your openai_api_key'

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0, api_key=api_key)
qa_chain = load_qa_chain(llm, chain_type="map_reduce")
qa_document_chain = AnalyzeDocumentChain(combine_docs_chain=qa_chain)
```

```python
answer = qa_document_chain.run(
    input_document=bbc_news,
    question="How much debt the country has?",
)

Markdown(answer)
```

```python
answer = qa_document_chain.run(
    input_document=bbc_news,
    question="Summarize me Sri Lanka's economic crisis situation",
)

Markdown(answer)
```
